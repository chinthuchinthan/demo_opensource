```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


%matplotlib inline
sns.set_style("whitegrid")
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
pd.set_option("display.float_format", "{:.2f}".format)
df.describe()
```

```python
df['SepalLengthCm'].value_counts().plot(kind='bar', color=['salmon', 'lightblue'])
```

```python
# checking for messing values
df.isnull().sum()
```

```python
categorical_val = []
continous_val =[]
for column in df.columns:
  if len(df[column].unique()) <= 10:
    categorical_val.append(column)
  else:
    continous_val.append(column)
```

```python
plt.figure(figsize=(15, 15))

for i, column in enumerate(categorical_val, 1):  # start index at 1
    plt.subplot(3, 3, i)

    df[df["SepalWidthCm"] == 0][column].hist(bins=35, color='blue',
                                       label='Have Heart Disease = NO', alpha=0.6)
    df[df["PetalWidthCm"] == 1][column].hist(bins=35, color='red',
                                       label='Have Heart Disease = YES', alpha=0.6)

    plt.legend()
    plt.xlabel(column)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df.SepalLengthCm[df.Species == 'Iris-virginica'],
            df.PetalLengthCm[df.Species == 'Iris-virginica'],
            c="salmon", label="Iris-virginica")
plt.scatter(df.SepalLengthCm[df.Species == 'Iris-setosa'],
            df.PetalLengthCm[df.Species == 'Iris-setosa'],
            c="lightblue", label="Iris-setosa")
plt.scatter(df.SepalLengthCm[df.Species == 'Iris-versicolor'],
            df.PetalLengthCm[df.Species == 'Iris-versicolor'],
            c="lightgreen", label="Iris-versicolor")
plt.title("Sepal Length vs Petal Length for Iris Species")
plt.xlabel("Sepal Length (Cm)")
plt.ylabel("Petal Length (Cm)")
plt.legend()
plt.show()
```

```python
corr_matrix = df.drop('Species', axis=1).corr()
fig,ax = plt.subplots(figsize=(15,15))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="YlGnBu");
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
